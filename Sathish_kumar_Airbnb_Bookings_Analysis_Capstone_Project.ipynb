```python
sns.countplot(da["neighbourhood_group"])
plt.title("List_of_Neighbourhood")
plt.show()
```

```python
sns.catplot(x='room_type', kind="count", hue="neighbourhood_group", data=data);
plt.title("differnt room types of neighbourhood_group")
plt.show()
# A majority of home/apt are located in Manhattan
# A majority of private rooms are located in Brooklyn
```

```python
room_prices=da['price'].value_counts()
room_prices.sort_index(inplace=True)
room_prices
```

```python
minimum_nights_count=da['minimum_nights'].value_counts()
minimum_nights_count.sort_index(inplace=True)
minimum_nights_count
```

```python
availability_of_room_in_a_year = da.groupby(["room_type"])["availability_365"].aggregate(np.mean).reset_index().sort_values("availability_365")
availability_of_room_in_a_year
sns.barplot(x="room_type", y="availability_365", data=da, order=availability_of_room_in_a_year["room_type"])
plt.show()
# Shared room has high availability in a year
```

```python
availability_of_room_in_a_year = da.groupby(["number_of_reviews","room_type","price"])["availability_365"].aggregate(np.mean).reset_index().sort_values("price")
availability_of_room_in_a_year
sns.set_palette("muted")
x = ('reviews_per_month')
y = ("availability_365")
title = 'availability relation to number of review per month'
data_filtered = da.loc[(da["availability_365"] < 365) & (da['reviews_per_month'] < 100)]
f, ax = plt.subplots(figsize=(8, 6))
sns.scatterplot(x=x,y=y,data=data_filtered)
plt.title(title)
plt.ioff()
```

```python
result = da.groupby(["room_type"])['price'].aggregate(np.median).reset_index().sort_values('price')
result
```

```python
result1 = da.groupby(["neighbourhood_group"])['price'].aggregate(np.median).reset_index().sort_values('price')
result1
```

```python
plt.figure(figsize=(10,9))
sns.barplot(x="room_type", y="price", data=da, order=result["room_type"])
plt.show()
```

```python
plt.figure(figsize=(15,9))
sns.barplot(x='neighbourhood_group', y="price", data=da, order=result1['neighbourhood_group'])
plt.show()
# Properties in Manhattan are more expensive
# The dataset can be separate between low price and high price properties
```

```python
price_of_rooms_on_hotels=da.groupby(['room_type',"neighbourhood_group"])["price"].count().unstack()
price_of_rooms_on_hotels
```

```python
price_of_rooms_on_hotels.plot()
plt.ylabel("price")
plt.show()
# shared price room is very low as compared with other rooms
# rooms of shared and private at Brooklyn is high as compared with other neighbourhood
# home/apt of Manhattan is high as compared with other neighbourhood
# Staten has the low price in all segments as compared with other neighbourhood
```
